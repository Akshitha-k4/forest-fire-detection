```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import Flatten
from keras. layers import MaxPooling2D
from keras.layers import Dense
import numpy as np
```

```python
model=Sequential()
model.add(Conv2D(32,(3,3),input_shape=(250,250,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
keras.layers.Conv2D(filters=256, kernel_size=(5,5), strides=(1,1), activation='relu', padding="same"),
model.add(Conv2D(34,(5,5),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(units=128,activation='relu'))
model.add(Dense(units=130,activation='relu'))
model.add(Dense(units=150,activation='relu'))
model.add(Dense(units=130,activation='relu'))
model.add(Dense(units=180,activation='relu'))
model.add(Dense(units=160,activation='relu'))
model.add(Dense(units=140,activation='relu'))
model.add(Dense(units=190,activation='relu'))
model.add(Dense(units=130,activation='relu'))
model.add(Dense(units=170,activation='relu'))
model.add(Dense(units=154,activation='relu'))
model.add(Dense(units=128,activation='relu'))
model.add(Dense(units=130,activation='relu'))
model.add(Dense(units=115,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.compile(optimizer='adadelta',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(rescale = 1./255,shear_range = 0.2,zoom_range = 0.4,horizontal_flip = True)
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
training_set =train_datagen.flow_from_directory('/content/drive/MyDrive/forest_fire/Training and Validation',target_size = (250,250),batch_size = 8,class_mode = 'binary')
test_set = test_datagen.flow_from_directory('/content/drive/MyDrive/forest_fire/Testing',target_size = (250,250),batch_size = 8,class_mode = 'binary')
```

```python
model.fit(training_set,steps_per_epoch = 10,epochs = 50)
```

```python
test_loss, test_acc = model.evaluate(test_set, verbose=2)

print("\n\n")
print("Test Loss: \t", test_loss, "\n")
print("Test accuracy: \t", test_acc, "\n")
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
def compute_gradients(model, input_image, target_class_idx):
    with tf.GradientTape() as tape:
        # Convert input_image to a tf.Tensor and watch it
        input_tensor = tf.convert_to_tensor(input_image)
        tape.watch(input_tensor)
        predictions = model(input_tensor)
        loss = predictions[:, target_class_idx]

    gradients = tape.gradient(loss, input_tensor)
    return gradients

def smoothgrad(model, input_image, target_class_idx, num_samples=50, noise_level=0.1):
    smooth_grad = np.zeros_like(input_image, dtype=np.float32)
    for i in range(num_samples):
        noise = noise_level * np.random.normal(loc=0.0, scale=1.0, size=input_image.shape)
        noisy_image = input_image + noise
        grad = compute_gradients(model, noisy_image, target_class_idx)
        smooth_grad += grad
    smooth_grad /= num_samples
    return smooth_grad

test_image, label = next(test_set)

# Select a single image from the batch for prediction
test_image = test_image[0]  # Select the first image (index 0)
test_image = test_image[np.newaxis, ...]  # Add batch dimension

predictions = model.predict(test_image)
target_class_idx = np.argmax(predictions[0])

smooth_grad_map = smoothgrad(model, test_image, target_class_idx, num_samples=50, noise_level=0.1)
smooth_grad_map = np.squeeze(smooth_grad_map)

smooth_grad_map = np.abs(smooth_grad_map).max(axis=-1)
smooth_grad_map = (smooth_grad_map - smooth_grad_map.min()) / (smooth_grad_map.max() - smooth_grad_map.min())

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(test_image[0])
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(smooth_grad_map, cmap='viridis')
plt.title("SmoothGrad Saliency Map")
plt.show()
```

```python
def occlusion_sensitivity(model, image, label, patch_size=8):
    width, height, _ = image.shape
    heatmap = np.zeros((width, height))
    for i in range(0, width, patch_size):
        for j in range(0, height, patch_size):
            occluded_image = image.copy()
            occluded_image[i:i+patch_size, j:j+patch_size] = 0
            occluded_image = np.expand_dims(occluded_image, axis=0)
            prediction = model.predict(occluded_image)
            heatmap[i:i+patch_size, j:j+patch_size] = prediction[0][label]
    heatmap = (heatmap - np.min(heatmap)) / (np.max(heatmap) - np.min(heatmap))
    return heatmap
```

```python
# Choose an image from the test set for explanation
from matplotlib import pyplot as plt
test_image, _ = next(test_set)
# Access the first image and add batch dimension
test_image = test_image[0][np.newaxis, ...]
test_label = np.argmax(model.predict(test_image)) # Predict on the correctly shaped input
# Compute Occlusion Sensitivity
occlusion_heatmap = occlusion_sensitivity(model, test_image[0], test_label)
plt.imshow(occlusion_heatmap, cmap='viridis')
plt.title("Occlusion Sensitivity")
plt.show()
```

```python
import cv2
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Fetch a batch of images from the test generator
test_images, _ = next(test_set)

# Define the Saliency Map function
def saliency_map(image, model, class_index):
    image = tf.convert_to_tensor(image, dtype=tf.float32)

    with tf.GradientTape() as tape:
        tape.watch(image)
        predictions = model(image, training=False)
        loss = predictions[:, class_index]

    grads = tape.gradient(loss, image)
    saliency = tf.reduce_max(tf.abs(grads), axis=-1)

    return saliency.numpy()

# Fetch a single image from the test data for explanation
test_image = test_images[0:1]  # Use the first image

# Predict the class for the test image
predictions = model.predict(test_image)
predicted_class = np.argmax(predictions)

# Compute Saliency Map
saliency_map_img = saliency_map(test_image, model, predicted_class)

# Enhance Saliency Map by applying Gaussian Blur
saliency_map_img = cv2.GaussianBlur(saliency_map_img, (5, 5), 0)

# Normalize the saliency map
saliency_map_img = np.squeeze(saliency_map_img)
saliency_map_img = (saliency_map_img - saliency_map_img.min()) / (saliency_map_img.max() - saliency_map_img.min())

# Convert to uint8 for visualization
saliency_map_img = np.uint8(255 * saliency_map_img)

# Overlay Saliency Map on Original Image
overlay = cv2.applyColorMap(saliency_map_img, cv2.COLORMAP_JET)
overlay = cv2.addWeighted(overlay, 0.5, cv2.cvtColor(np.uint8(255 * test_image[0]), cv2.COLOR_RGB2BGR), 0.5, 0)

# Visualize the Original Image, Saliency Map, and Overlay
def plot_saliency(image, saliency_map, overlay):
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.title('Original Image')
    plt.imshow(image[0])
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.title('Saliency Map')
    plt.imshow(saliency_map, cmap='hot')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.title('Overlay of Saliency Map on Original Image')
    plt.imshow(cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.show()

plot_saliency(test_image, saliency_map_img, overlay)
```
